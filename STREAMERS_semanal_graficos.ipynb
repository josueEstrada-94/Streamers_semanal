```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%config IPCompleter.greedy = True

pd.options.display.float_format = '{:15.2f}'.format
```

```python
ds_sem = pd.read_excel('/content/Libro1 10 al 16-04.xlsx')
```

```python
ds_sem.rename(columns=lambda x: x.replace(' ', '_'), inplace=True)
ds_sem.rename(columns=lambda x: x.replace('\n', ''), inplace=True)
```

```python
ds_sem.rename(columns= {'ID':'id'}, inplace = True)
```

```python
ds_sem.set_index('id', inplace = True) 
```

```python
def no_trabajo(row):
    coins_total = row['Total_coins10/04-16/04']
    category = row['Nivel']
    if (coins_total == 0) | (category == 'UNKNOWN'):
        return ('No Trabajo')
    else:
        return ('Trabajo')
```

```python
ds_sem ['Trabajo_semanal'] = ds_sem.apply(no_trabajo, axis = 1)
```

```python
ds_sem['Total_coins10/04-16/04'].describe()
```

```python
ds_sem.info()
```

```python
def no_trabajo_mes(row):
    coins_total = row['Final_Coins(Ultimas_4_semanas)']
    category = row['Nivel']
    if (coins_total == 0) | (category == 'UNKNOWN'):
        return ('No Trabajo')
    else:
        return ('Trabajo')
```

```python
ds_sem ['Trabajo_mensual'] = ds_sem.apply(no_trabajo_mes, axis = 1)
```

```python
ds_sem.Trabajo_mensual.value_counts()
```

```python
def coins_por_semana(id):
    
    registro = ds_sem[ds_sem.index == id]
    
    if not registro.empty:
        nombre = registro['User_Name'].values[0]
        trabajo_semanal = registro['Trabajo_semanal'].values[0]
        
        if trabajo_semanal == 'Trabajo':
            dinero_acumulado = registro['Total_coins10/04-16/04'].values[0]
            
            plt.bar(nombre, dinero_acumulado)
            plt.ylabel('Coins Acumulados')
            plt.title('Total de Coins Generados en una Semana')
            plt.ylim(0, 220000)
            plt.show();
        elif trabajo_semanal == 'No Trabajo':
            print('La usuaria no gener√≥ dinero en la semana del 10 - 16 de Abril.')
        
    else:
        print('N√∫mero de registro no encontrado')

num_registro = input('Ingrese el N√∫mero de registro: ')
coins_por_semana(int(num_registro))
```
